```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.array([1,3,6,9])
#logarithm
y = np.log(x) # also we use sin(),tan() and other functions
plt.plot(x,y)
plt.title("logarithm chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#default x value -> 0,1,2,3.... same size with y 
y = np.random.rand(5)

plt.plot(y,"o:r") 
plt.show()
```

```python
x =np.random.rand(40)
y = np.random.rand(40)
colors = np.random.randint(0,100,40)
sizes = np.random.randint(100,200,40)

plt.scatter(x,y,c=colors,s=sizes)
plt.title("Point Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = ['A','B','C']
y = np.random.randint(0,10,3)

plt.bar(x,y)
plt.title("Colum chart")
plt.xlabel("Categories")
plt.ylabel("Values")    
plt.show()
```

```python
data = np.random.randn(500)
plt.hist(data,bins=20)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
x = np.array([10,6,12,20])
mylabels =np.array(['a','b','c','d'])
plt.pie(x, labels=mylabels)
plt.title("Pie chart")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-4,4,2)
y = np.arange(-4,4,2)
x,y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(x,y,z)
ax.set_title("3D Graphics")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```
